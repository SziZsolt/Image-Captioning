```python
import pandas as pd
import numpy as np
import os
from keras import Model
from keras.applications import VGG16
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
images_directory = os.path.join('flickr30k_images', 'flickr30k_images')
```

```python
image_files = os.listdir(images_directory)
image_files = image_files[:3000]
image_files_count = len(image_files)
image_files
```

```python
feature_extractor = VGG16()
model = Model(inputs = feature_extractor.inputs, outputs = feature_extractor.layers[-2].output)
```

```python
df = pd.DataFrame(columns=['image_name','image_array', 'features'])
```

```python
for i, image_file in tqdm(enumerate(image_files)):
    img = Image.open(os.path.join(images_directory, image_file))
    img = img.resize((224,224)) 
    img = np.array(img)
    img = np.expand_dims(img, axis=0)
    img = img / 127.5
    img = img -1
    # pred = model.predict(img)[0]
    df.loc[i] = [image_file,img, None]
```

```python
def pred(row):
    return model.predict(row['image_array'])[0]
```

```python
df['features'] = df.apply(pred, axis=1)
```

```python
df
```

```python
df['features_list'] = df['features'].apply(lambda x: x.tolist())
```

```python
df_to_write = df[['image_name', 'features_list']].copy()
df_to_write.to_csv('features.csv')
```
