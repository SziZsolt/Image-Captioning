```python
import pandas as pd
import numpy as np
import tensorflow as tf
import keras
from keras.layers import TextVectorization
import ast
from nltk.tokenize import word_tokenize
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df['features_list'] = df['features_list'].apply(lambda x: ast.literal_eval(x))
df['captions'] = df['captions'].apply(lambda x: ast.literal_eval(x))
```

```python
all_captions = []
max_length = 0
for i, row in df.iterrows():
    for caption in row['captions']:
        splitted = caption.split()
        all_captions.append(caption)
        length = len(splitted)
        if length > max_length:
            max_length = length
```

```python
text_dataset = tf.data.Dataset.from_tensor_slices(all_captions)
```

```python
text_vectorization = TextVectorization(output_mode='int')
text_vectorization.adapt(text_dataset)
```

```python
vocab = text_vectorization.get_vocabulary()
vocab_size = text_vectorization.vocabulary_size()
vocab
```

```python
def tokenize_captions(row):
    tokenized_captions= []
    for caption in row['captions']:
        tokenized_caption = text_vectorization(caption).numpy().tolist()
        tokenized_captions.append(tokenized_caption)
    return tokenized_captions
```

```python
df['tokenized_captions'] = df.apply(tokenize_captions, axis=1)
```
